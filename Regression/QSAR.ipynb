# Machine Learning Model

## Linear Regression

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat,df_target, test_size = 0.3, random_state=42)
```

```python
X_train.head()
```

```python
#Import model
from sklearn.linear_model import LinearRegression
```

```python
#Assign Model
model = LinearRegression()
```

```python
#Train model
model.fit(X_train, y_train)
```

```python
print(model.intercept_)
```

```python
model.coef_
```

```python
X_train.columns
```

```python
coef = pd.DataFrame(data = model.coef_[0],
                   index = X_train.columns,
                   columns=["Coeficiente corelacin"])
```

```python
coef
```

```python
model.score
```

# Predictions

Write a function that returns descriptors for a single compound

```python
def test_compound(ID):
    test = Data[Data["Compound's ID"]== ID]
    test = test[feature_names]
    return test
```

Writte a function to look for an specific compound descriptors

```python
test = test_compound(1603)
test
```

```python
#Single prediction
model.predict(test)
```

Figure out real compound's activity

```python
def test_compound_Activity(ID):
    test = Data[Data["Compound's ID"]== ID]
    Activity = test["Activity"]
    return Activity
```

```python
activity = test_compound_Activity(1603)
activity
```

```python
def report(ID, test, activity):
    print("Evaluation of compund ID", str(ID))
    print("Predicted activity value: ", str(model.predict(test)[0][0]))
    print("Real activity value", str(activity))
```

```python
report(1603, test, activity)
```

# Evaluate model

Note: Regression Metrics

scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, predictions)
```

```python
#Compute R2
print("R^2:",r2_score(y_test, predictions))
```

```python
#Compute metrics
print("MAE:", metrics.mean_absolute_error(y_test, predictions))
print("MSE:", metrics.mean_squared_error(y_test, predictions))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
