# Create a DataFrame

```python
antidiabeticos = ["Metformina", "Glimepirida"," Empagliflozina"]
```

```python
dosis_antidiabeticos = [500, 4, 12]
```

```python
antihipertensivos = ["Atenolol", "Captopril", "Enalapril"]
```

```python
dosis_antihipertensivos = [10,50, 20 ]
```

merge data into a list

```python
data = [antidiabeticos, dosis_antidiabeticos, antihipertensivos, dosis_antihipertensivos]
```

```python
data = np.transpose(data, axes = None)
```

write Data Frame

```python
DF1 = pd.DataFrame( data = data,
                    columns = ["antidiabeticos", "dosis", "antihipertensivo", "dosis"])
```

Data frame from a dictionary

```python
data_dict = {
    "antidiabeticos":antidiabeticos,
    "dosis antidiabeticos": dosis_antidiabeticos,
    "antihipertensivos": antihipertensivos,
    "dosis antihipertensivos": dosis_antihipertensivos
}
```

```python
#data_dict["antidiabeticos"]
```

```python
DF2 = pd.DataFrame.from_dict(data_dict)
```

```python
DF2
```

Create a new column

```python
DF2["antirritmicos"] = ["Metoprolol", "Amiodarona", "Diltiazem" ]
```

```python
DF2
```

save a data frame as csv file

```python
DF2.to_csv("MyfirstDF.csv", sep = ",")
```
